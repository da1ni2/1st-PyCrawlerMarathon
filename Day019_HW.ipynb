```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
html = BeautifulSoup(html_source , 'html.parser')
```

```python
tag = html.find('div' , class_ = 'ResultPanel').find('table' , class_ = 'TABLE_G').find('tbody').find_all('tr')
```

```python
tag_num = 1
gas = []
date = []
value = []

for i in tag[1:]:
    if ((tag_num - 1) % 11) == 0:
        gas.append(i.find('td').get_text())
        date.append(i.find_all('td')[2].get_text())
        value.append(float(i.find_all('td')[3].get_text()))
    else:
        date.append(i.find_all('td')[0].text)
        value.append(float(i.find_all('td')[1].text))
    tag_num += 1
        
```

```python
dict__ = {}
date = date[0:11]
s = 0
for i in gas:
    dict_ = {}
    for j in date:
        dict_[j] = value[s]
        s = s+1
    dict__[i] = dict_
```
