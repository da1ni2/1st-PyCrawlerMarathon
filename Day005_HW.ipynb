```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
#30筆
'''id
title
excerpt
anonymousSchool
anonymousDepartment
pinned
forumId
replyId
createdAt
updatedAt
commentCount
likeCount
withNickname
tags
topics
meta
forumName
forumAlias
gender
school
department
replyTitle
mediaMeta
reactions
hidden
customStyle
isSuspiciousAccount
layout
withImages
withVideos
media
reportReasonText
postAvatar'''


import requests
response = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
import json
data = json.loads(response.text)
len(data)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''
n=1
for i in data:
    print(f"第{n}筆資料")
    print(f"標題： {i['title']}")
    print(f"貼文時間： {i['createdAt']}")
    print(f"留言人數： {i['commentCount']}")
    print(f"按讚人數： {i['likeCount']}")
    print('\n')
    n=n+1
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

'''
Your Code
'''
import numpy as np

#熱門
comment_count = []
like_count = []

for i in data:
    comment_count.append(i['commentCount'])
    like_count.append(i['likeCount'])
print(f'熱們平均留言人數： {np.average(comment_count)}')
print(f'熱們平均按讚人數： {np.average(like_count)}')
    
```

```python
#非熱門
response = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data = json.loads(response.text)
comment_count = []
like_count = []

for i in data:
    comment_count.append(i['commentCount'])
    like_count.append(i['likeCount'])
print(f'非熱門平均留言人數： {np.average(comment_count)}')
print(f'非熱門平均按讚人數： {np.average(like_count)}')
```
