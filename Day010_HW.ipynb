```python
def sort_page():
    html = chrome.page_source
    html = BeautifulSoup(html , 'html5lib')
    article_tag = html.find_all('div' , class_ = 'r-ent')
    
    for i in article_tag:
        
        try:
            title = i.find('div' , class_ = 'title').text.replace('\n' , '').replace('\t','')
            link = 'https://www.ptt.cc' + i.find('div' , class_ = 'title').find('a')['href']
            date = i.find('div',class_='meta').find('div',class_='date').text
            today = time.strftime("%m/%d")
    
            r = requests.get(link,cookies = {'over18':'1'})
            html_content = BeautifulSoup(r.text , 'html5lib')
            content = html_content.find('div' , attrs = {"id":"main-content"}).get_text()
            content_text = ''
            for i in content:
                if i != '※':
                    content_text += i
                else:
                    break        
    
            if date == today:
                print(title)
                print(content_text)
                print('\n')
        except:
            continue
    return date 
            
```

```python
continue_ = True
while continue_:
    date = sort_page()
    yesterday = str(time.strptime(str(datetime.datetime.today() + timedelta(-1)) , '%Y-%m-%d %H:%M:%S.%f')[1]) + '/' + str(time.strptime(str(datetime.datetime.today() + timedelta(-1)) , '%Y-%m-%d %H:%M:%S.%f')[2])
    if date == yesterday:
        continue_ = False
    else:
        chrome.find_element_by_link_text("‹ 上頁").click()

    
```
